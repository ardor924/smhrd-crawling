```python
#================================================================
# ▶ 모듈 불러오기
#================================================================
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 1. url요청

# 브라우저 예외 : 가끔 특정사이트에서 406 에러발생
# 요청환경이 브라우저가 아닌 코드라는 이유로 거절(차단)되는것이 이유
# user-agent를 헤더에 넣어서 보내줘야 원하는 사이트에 접속이 가능
url = 'https://www.melon.com/chart/'
data =  \
{   
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36'
}
res = req.get(url,headers=data)
res
```

```python
# 결과를 파싱
soup = bs(res.text,"lxml")
```

```python
# 선택자로 a 태그 가져오기
cnt = 0
title_list = soup.select('div.ellipsis.rank01> span>a')
for title in title_list:
    cnt +=1
    print(f'{cnt} {title.text}',end='\n\n')
```
